```python
plt.figure(figsize=(6,7))
sns.histplot(df["Age"], bins=30,kde=True, color="blue")
plt.title("Employee Age Distribution")
plt.xlabel("age")
plt.ylabel("Number of Employees")
plt.show()
```

```python
plt.figure(figsize=(8,6))
df["EmployeeType"].value_counts().plot.pie(autopct="%1.1f%%", cmap="coolwarm", shadow=True)
plt.title("Percentage of Types of Hiring")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="DepartmentType", y="Engagement Score", data=df, palette="viridis")
plt.xticks(rotation=45)
plt.title("Average Participation Score by Section")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="Training Outcome", y="Engagement Score", data=df, palette="coolwarm")
plt.title("Impact of training on degree of participation")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = df[["Engagement Score", "Satisfaction Score", "Work-Life Balance Score", "Training Cost", "Training Duration(Days)"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Analysis of Different Factors")
plt.show()
```

```python
category_pivot = df.groupby('DepartmentType')['Current Employee Rating'].mean()

plt.figure(figsize=(12, 6))
sns.barplot(x=category_pivot.index, y=category_pivot.values)
plt.xticks(rotation=90)
plt.title('Average Employee Rating by Department')
plt.xlabel('Department')
plt.ylabel('Average Rating')
plt.show()
```

```python
categorical_cols =['BusinessUnit','EmployeeStatus',	'EmployeeType',	'PayZone',	'EmployeeClassificationType','RaceDesc','MaritalDesc','Training Program Name',	'Training Type'	,'Training Outcome']

for col in categorical_cols:
  plt.figure(figsize=(6,6))
  values = df[col].value_counts()
  plt.pie(values, labels=values.index, autopct='%1.1ff%%', shadow=True)
  plt.title(f'{col} Proportionality')
  plt.show()
```

```python
def plot_bar_pie(column, dataframe= df, top_n=5):
    """
    Plot barchart and piechart side by side.
    """
    # Check to ensure the column is categorical.
    if dataframe[column].dtypes == "O":
        dist = dataframe[column].value_counts(normalize=True).head(top_n)
        # Start the plotting the barchart.
        fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15, 5))
        ax0.bar(dist.index, dist.values)
        ax0.set_xticklabels(rotation=45, labels=dist.index)
        ax0.set_title(f"Bar chart Distribution for {column}")
        ax0.set_xlabel(column)
        ax0.set_ylabel("Frequency")
        # Plotting the pie chart
        ax1.pie(dist, labels=dist.index, autopct="%1.2f%%")
        ax1.set_title(f"Pie Chart Distribution for {column}")
        return fig.show()
    else:
        print("The column given is not a Categorical Column")
```

```python
plot_bar_pie("Title")
```
